```python
import pandas as pd

df= pd.read_csv("weather_data.csv", parse_dates=['day'])
df
```

```python
df.set_index('day', inplace=True)
df
```

```python
df.fillna(0)
df
```

```python
df.fillna({
    'temperature': df.temperature.mean(),
    'windspeed': df.windspeed.mean(),
    'event': 'No event'
})
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='bfill')
```

```python
df.interpolate()
```

```python
df.dropna()
```
