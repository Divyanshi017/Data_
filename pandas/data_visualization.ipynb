```python
import pandas as pd
```

```python
df_sales=pd.read_excel("linechart.xlsx")
df_sales.head()
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df_sales["Quarter"],df_sales["Fridge"],color="green", label="Fridge")
plt.plot(df_sales["Quarter"],df_sales["Dishwasher"],color="red", label="Dishwasher")
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"],color="black", label="Washing Machine")
plt.title("Product Sales")
plt.ylabel("Revenue($)")
plt.xlabel("Financial Quarter")
plt.legend()
```

```python
total_sales=df_sales[["Fridge","Dishwasher","Washing Machine"]].sum()
total_sales
```

```python
total_sales.index
```

```python
plt.pie(total_sales, labels=total_sales.index, autopct="%1.2f%%",explode=(0.1,0,0),shadow=True,startangle=50)
plt.show()
```

```python
df_sales.plot(kind="bar", x="Quarter")
plt.xticks(rotation=45)
plt.show()
```

```python
df_sales2=df_sales.set_index("Quarter")
df_sales2
```
