```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
```

```python
df=pd.DataFrame({
    "area": [200,250,300,350,400],
    "price": [1000,3000,5000,7000,9000]
})
df
```

```python
%matplotlib inline 
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price, color='red')
df.to_csv("file.csv", index=False)
```

```python
df1=pd.read_csv('file.csv')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df1[['area']],df1.price)
```

```python
reg.predict([[300]])
```

```python
reg.coef_
```

```python
np.round(reg.intercept_)
```
